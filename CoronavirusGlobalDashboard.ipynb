## Preprocessing

```python
def prepare_data(df):
    df = df.drop(columns=['Lat', 'Long', 'Province/State'])
    df_grouped = df.groupby(['Country/Region'], as_index=False).sum()
    df_grouped = df_grouped.set_index('Country/Region').transpose()
    df_grouped.reset_index(level=0, inplace=True)
    df_grouped.rename(columns={'index': 'Date'}, inplace=True)
    df_grouped['Date'] = pd.to_datetime(df_grouped['Date'])
    return df_grouped
```

## Removing Outliers

```python
import numpy as np
def Remove_outliers(df,col):
    for i in col:
        q75, q25 = np.percentile(df.loc[:,i], [75, 25])
        iqr = q75 - q25
        Innerfence = q25 - (iqr*1.5)
        Upperfence = q75 + (iqr*1.5)
    
        # replace outliers with NA

        df.loc[df[i]<Innerfence, i] = np.nan
        df.loc[df[i]>Upperfence, i] = np.nan
        #print(df.isnull().sum())
        
        df[i] = df[i].fillna(df[i].median())

    return df
```

# APP

```python
app = dash.Dash(__name__,external_stylesheets=['https://codepen.io/chriddyp/pen/bWLwgP.css'])
```

```python
colors = {
    'background': '#f2f2f2',
    'text': '#E1E2E5',
    'figure_text': '#ffffff',
    'confirmed_text':'#3CA4FF',
    'deaths_text':'#f44336',
    'recovered_text':'#5A9E6F',
    'highest_case_bg':'#393939',
    
}

divBorderStyle = {
    'backgroundColor' : '#393939',
    'borderRadius': '12px',
    'lineHeight': 0.9,
}

boxBorderStyle = {
    'borderColor' : '#393939',
    'borderStyle': 'solid',
    'borderRadius': '10px',
    'borderWidth':2,
}
```

```python
app.layout= html.Div([
    # Add a title
    html.H1("Coronavirus Global Dashboard",style={'textAlign': 'center'}),
    
    # ADD the BANS
    html.Div([
            html.Div([
                html.H4(children='Total Cases',
                       style={
                           'textAlign': 'center',
                           'color': colors['confirmed_text'],
                       }
                       ),
                html.P(f"{df_confirmed_total[-1]:,d}",
                       style={
                    'textAlign': 'center',
                    'color': colors['confirmed_text'],
                    'fontSize': 30,
                }
                ),
        
            ],
                style=divBorderStyle,
                className='four columns',
            ),
            html.Div([
                html.H4(children='Total Deceased',
                       style={
                           'textAlign': 'center',
                           'color': colors['deaths_text'],
                       }
                       ),
                html.P(f"{df_deaths_total[-1]:,d}",
                       style={
                    'textAlign': 'center',
                    'color': colors['deaths_text'],
                    'fontSize': 30,
                }
                ),
                
            ],
                style=divBorderStyle,
                className='four columns'),
            html.Div([
                html.H4(children='Total Recovered',
                       style={
                           'textAlign': 'center',
                           'color': colors['recovered_text'],
                       }
                       ),
                html.P(f"{df_recovered_total[-1]:,d}",
                       style={
                    'textAlign': 'center',
                    'color': colors['recovered_text'],
                    'fontSize': 30,
                }
                ),
                
            ],
                style=divBorderStyle,
                className='four columns'),
        ], className='row'),
     html.H3("Death cases around the world", style={'textAlign': 'center'}),
    # Adding The figures
     html.Div([
         dbc.Row(
            [
             # Map
             dbc.Col(
             html.Div(
                   
                   dcc.Graph(id='map',figure=geo) 
               ), style={'display': 'flex', 
                  'flexDirection': 'column',
                  'width': '100%'}
            ),
          
          dbc.Col(
                # Bar Plot
                html.Div([
                html.H3("Recovered cases per country", style={'textAlign': 'center'}),
                dcc.Graph(id='bar',figure=bar),
                ])
              , style={'display': 'flex', 
                  'flexDirection': 'column',
                  'width': '100%'}
            )   
            ]),
          dbc.Row(
              # Line
             html.Div(
               [         
                   html.H3("Confirmed cases accross time", style={'textAlign': 'center'}),
                   dcc.Dropdown(id='input-country',placeholder="Select a Country",
            options=[{'label': i, 'value': i} for i in dropdown_deathes]
                     ) , 
                dcc.Graph(id='line',figure=line)
               ])
        )
     ])
    
    
])
@app.callback(
    Output('line','figure'),
    Input('input-country','value'))
  
def update_line(value):
    if value==None:
        line = px.line(df_without_con, x="Date", y="Egypt")
        return line
    else:
        line = px.line(df_without_con, x="Date", y=value)
        return line
```
